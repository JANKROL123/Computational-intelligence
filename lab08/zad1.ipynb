### Import

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import matplotlib.pyplot as plt 
f = open("article.txt", "rt").read()
```

### Tokenizacja

```python
nltk.download("punkt")
tokens = nltk.word_tokenize(f)
print(tokens)
print(len(tokens))
```

### Stop words

```python
nltk.download("stopwords")
stop_words = set(stopwords.words("english"))
filtered_tokens = []
for word in tokens:
    if word not in stop_words:
        filtered_tokens.append(word)
tokens = filtered_tokens
print(tokens)
print(len(tokens))
```

### Lematyzacja

```python
lemmatizer = WordNetLemmatizer()
nltk.download('wordnet')
lemetized_tokens = [lemmatizer.lemmatize(word) for word in tokens]
tokens = lemetized_tokens
print(tokens)
print(len(tokens))
```

### Usunięcie innych niepotrzebnych znaków

```python
filtered = []
for word in tokens:
    if word.isalnum():
        filtered.append(word)
tokens = filtered
print(tokens)
print(len(tokens))
```

### Wektor

```python
freq_dist = nltk.FreqDist(tokens)
top = dict(freq_dist.most_common(10))
plt.bar(top.keys(), top.values())
plt.title('Top 10')
plt.xlabel('Words')
plt.ylabel('Occurances')
plt.xticks(rotation=20)
plt.show()
```
